```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, matthews_corrcoef, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load actual results
actual_df = pd.read_csv("sample_updated.csv")
actual_labels = actual_df["Label"]

# Load predicted results
predicted_df = pd.read_csv("2025-08-18T12-21_export.csv")
# Convert text labels to 0 and 1 (example: 'Positive'->1, 'Negative'->0)
predicted_labels = predicted_df["Prediction"].replace({"DNA Binding": 1, "Non-DNA Binding": 0})

# Generate confusion matrix
cm = confusion_matrix(actual_labels, predicted_labels)
print(accuracy_score(actual_labels, predicted_labels))
print(matthews_corrcoef(actual_labels, predicted_labels))
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```
